```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
data = [[8,8,'cat'],[9,12,'dog'],[15,15,'dog'],[7,9.5,'cat'],[20,18,'dog']]
df = pd.DataFrame(data,columns = ["weight","height","label"])
print(df)
X = df[["weight","height"]]
Y = df["label"]
tree = DecisionTreeClassifier()
mod = tree.fit(X,Y)
```

```python
predictions = tree.predict([[26,20]])
```

```python
from sklearn.externals.six import StringIO
from sklearn.tree import export_graphviz
import pydotplus
import os
os.environ["PATH"]+= os.environ["PATH"]+';'+r'C:\ProgramData\Anaconda3\Library\bin\graphviz2.38\bin'
from IPython.display import Image
dot_data = StringIO()
export_graphviz(mod,out_file = dot_data,filled = True ,rounded = True,feature_names = X.columns,
                class_names = ["cat","dog"])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
